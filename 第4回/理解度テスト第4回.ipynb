**Q2 GANと暗黙的生成モデルの説明について最も適切なものを選択せよ**
1. 暗黙的生成モデルでは初めに尤度関数を定義し、これを最適化することで生成器を学習する
2. 尤度が計算できないのでエビデンス下界を最大化する
3. GANの学習はデータ分布とモデル分布のKLダイバージェンス最小化に対応している
4. 尤度が計算できないので、識別器は密度比を直接求める

**Q3 一般的なGANの説明に関して最も当てはまるものを選択せよ**

3. 勾配消失問題が起こる主な原因は、誤差が識別器と生成器の両方を通じて伝搬されるため、層が深くなりすぎることである
4. GANの学習は識別器と生成器の間のミニマックスゲームとして解釈できる
